This is a short program counting the number of times a word appears on a webpage.

Start by installing the library requests, it scrapes data from a webpage

```python
import requests
```

identify the URL you intend to use and requests scrapes the webpage.

```python
URL = 'https://en.wikipedia.org/wiki/Astronomy'
page = requests.get(URL)
print(page.text)
```

install html2text, converts html to easily readable text

```python
!pip install html2text
```

```python
import html2text
```

calls the function HTML2TEXT from the library html2text
ignores the links in the text

```python
 h = html2text.HTML2Text()
 h.ignore_links = True
```

extracts the text from the data

```python
text=h.handle(page.text)
```

split the data to create a list

```python
words=text.split()
print(words)
```

created a list Sun and appended the initially created list into it

```python

Sun=[]
for word in words:
    Sun.append(word)
print(Sun)
```

created a list with punctuations marks and a for loop to go through the two lists and remove any items in the list Sun attached to any punctuation marks,this was to farther clean the data and remove the remaining links in the data.

```python
punctuations=['.','!',';','()','[]','_','#']
for item in punctuations:
    for items in Sun:
      if item in items:
        Sun.remove(items)

```

created a list unique and appended all the unique words to it

```python
unique=[]
for items in Sun:
  if items not in unique:
    unique.append(items)
print(len(unique))
```

created a list sorted where all the unique words in the list Sun and the number of times they appear are recorded.

```python
Sorted=[]
for items in unique:
  i=0
  for item in Sun:
    if item==items:
      i=i+1
  Sorted.append(items+'-'+str(i))
```

```python
print(Sorted)
```
